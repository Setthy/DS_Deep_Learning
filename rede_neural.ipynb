# Rede Neural na prática.

```python
# importando as bibliotecas:
    
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
```

```python
# Carregamento da base de dados e criação dos previsores (variáveis independentes - X) e da classe ( variável dependente - y):

base = datasets.load_iris()
previsores = base.data
classe = base.target
classe
```

```python
# Transformação da classe para o formato 'dummy', pois tomes uma rede neural com 3 neurônios na camada de saída:

classe_dummy = np_utils.to_categorical(classe)
classe_dummy
```

```python
# Divisão da base de dados entre trinamento e teste (30% para testar e 70% para treinar)

X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores, classe_dummy, test_size = 0.3, random_state = 0)
```

```python
# Criação da estrutura da rede neural com classe Sequential (sequência de camadas):

modelo = Sequential()

# Primeira camada oculta, 5 neuronios, 4 neuronios de entrada:

modelo.add(Dense(units = 5, input_dim = 4))

# Segunda camada oculta:

modelo.add(Dense(units = 4))

# Terceira camada oculta:

# Função sotmax porque temos um problema de classificação com mias de duas classes é gerado uma probabilidade em cada neuronio:

modelo.add(Dense(units = 3, activation = 'softmax'))
```

```python
# Visualização da estrutura de rede neural:

modelo.summary()
```

```python
# Configuração dos par{ametros da rede neural (adam = algoritimo para atualizar os pesos e loss = cálculo do erro):

modelo.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Treinamento, dividindo a base de treinamento em uma porção para validação (validation_data):

modelo.fit(X_treinamento, y_treinamento, epochs = 1000, validation_data = (X_teste, y_teste))
```

```python
# Previsões e mudar a variável para True ou False de acordo com threshold 0.5:

previsoes = modelo.predict(X_teste)
previsoes = (previsoes > 0.5)
previsoes
```

```python
# Como é um problema com três saidas, precisamos buscar a posição que possui o maior valor (são retornados 3 valores)

y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsoes_matrix = [np.argmax(t) for t in previsoes]
```

```python
# Geração da matriz de confusão:

confusao = confusion_matrix(y_teste_matrix, y_previsoes_matrix)
confusao
```
